Insight:import libraries and load dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

df.head()
```

Insight:The dataset has 891 rows and 15 columns. Features like age, fare, and embarked contain missing values.

```python
df.shape
df.info()
df.describe()
```

Insight:Columns age, embarked, deck have missing data. We'll need to impute or drop them for modeling.

```python
df.isnull().sum()
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

Insight:About 38% of passengers survived. Most passengers were in the 20–40 age range.

```python
sns.countplot(x='survived', data=df)
sns.histplot(df['age'].dropna(), kde=True)
```

Insight:Females had a much higher survival rate than males. First-class passengers were more likely to survive than those in third class.

```python
sns.countplot(x='sex', hue='survived', data=df)
sns.countplot(x='class', hue='survived', data=df)
```
